```python
df = pd.DataFrame({'A': [1, 2, 3, 4, 5, 6, 7, 8, 9]})

# define the column name and three consecutive values
col = 'A'
consecutive_values = [3, 4, 5]

# get the index of the first occurrence of the consecutive values
start_index = df.loc[df[col].eq(consecutive_values[0])].index

# initialize a list to store the rows corresponding to consecutive values
rows = []

# iterate over the start indexes and check if the next two values are consecutive
for idx in start_index:
    if df.loc[idx:idx+len(consecutive_values)-1, col].tolist() == consecutive_values:
        rows.extend(range(idx, idx+len(consecutive_values)))

# use the row numbers to filter the rows of the DataFrame
result = df.iloc[rows]
start_index
```

```python
data_file = pd.DataFrame(df['2010-01-01':'2020-12-31']['belgrade_water_level_cm'])
data_file.head()
```

```python
n_steps = 4

# split into train and test sets
train_size = int(len(data_file) * 0.75)
test_size = len(data_file) - train_size
train_file, test_file = data_file[0:train_size], data_file[train_size : len(data_file)]
print(test_file)

print('Length of training set: ',len(train_file), 'Length test set: ', len(test_file))

# split a univariate sequence into samples
def split_sequence(sequence, n_steps):
    X, y = list(), list()
    for i in range(len(sequence)):
 # find the end of this pattern
        end_ix = i + n_steps
 # check if we are beyond the sequence
        if end_ix > len(sequence)-1:
            break
 # gather input and output parts of the pattern
        seq_x, seq_y = sequence[i:end_ix], sequence[end_ix]
        X.append(seq_x)
        y.append(seq_y)

    return array(X), array(y)

trainX, trainY = split_sequence(train_file['belgrade_water_level_cm'], n_steps)
testX, testY   = split_sequence(test_file['belgrade_water_level_cm'], n_steps)

print('hello there!')
 
    
testX_df = pd.DataFrame (testX, columns = ['lag{}'.format(i+1) for i in range(n_steps)])
testY_df = pd.DataFrame(testY)
testY_df.columns = ['Values to be predicted']
test_total = pd.concat([testX_df, testY_df], axis=1)
test_total.head(15)
```
